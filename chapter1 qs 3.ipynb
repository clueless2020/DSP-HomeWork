## Libraries required

```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import warnings
warnings.filterwarnings('ignore')

from IPython.html.widgets import interact, fixed
from IPython.display import display
import matplotlib.pyplot as plt
#import librosa
```

## generaing a SinSignal

```python

x2= thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
```

## Displaying sine signal

```python
x2.plot()
```

## Generating a cosSignal

```python
x3= thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
```

## Displaying cos signal

```python
x3.plot()
```

## Adding sin signal and cos signal

```python
mix=x2+x3
```

## Converting ignals to wave

```python

wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
from IPython.display import Audio
audio = Audio(data=wave.ys, rate=wave.framerate)

wave.make_audio()
```

```python
wave.plot()
```

```python
w1_seg= wave.segment(start=0.1, duration=.02)
w1_seg.plot()
```

## Calculating spectrum

```python
spectrum = w1_seg.make_spectrum()
```

```python
spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spectrum.plot(high=10000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spectrum.low_pass(3000) #attenuates all frequency components above a cutoff frequency.
```

```python
filtered = spectrum.make_wave()
```

```python
filtered.normalize()
```

```python
filtered.apodize()
filtered.plot()
thinkplot.config(xlabel='Time (s)')
```

## Filtered audio

```python

filtered.make_audio()
```
