## Question 2 Part 1 : Make a square signal at 1100 Hz and make a wave that samples it at 10000 frames per second.

## Constructing a Square Signal

```python
square = thinkdsp.SquareSignal(1100).make_wave(duration=10, framerate=10000)
```

## Ploting the generated Square Signal

```python
segment = square.segment(duration=0.005)
segment.plot()
```

## Convert the generated square signal to audio

Purpose: To play the generated square signal

```python
square.make_audio()
```

## Calculating Spectrum and plotting generated Square Signal

```python
square.make_spectrum().plot()
```
